```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
sns.boxplot(x= df['cgpa'])
plt.show()
```

```python
q1 = df['cgpa'].quantile(0.25)
q3 = df['cgpa'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr 
higher = q3 + 1.5 * iqr

df = df[ (df['cgpa'] <= higher ) & (df['cgpa'] >= lower)]

sns.boxplot(x = df['cgpa'])
plt.show()
```

```python
sns.boxplot(x = df['placement_exam_marks'])
plt.show()
```

```python
q1 = df['placement_exam_marks'].quantile(0.25)
q3 = df['placement_exam_marks'].quantile(0.75)
iqr = q3 - q1 
lower = q1 - 1.5 * iqr 
higher = q3 + 1.5 * iqr 

df = df[ (df['placement_exam_marks'] <= higher ) & (df['placement_exam_marks'] >= lower)]
```

```python
sns.boxplot(x = df['placement_exam_marks'])
plt.show()
```

```python
new_df = df.copy()

new_df['placement_exam_marks'] = np.where(
  new_df['placement_exam_marks']  > higher, higher,
  np.where(
    new_df['placement_exam_marks'] < lower,lower,new_df['placement_exam_marks']
  )
)
```
