```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
df.head(3), df.shape
```

```python
df['Height'].describe()
```

```python
sns.distplot(df['Height'])
# plt.show()
```

```python
sns.boxplot(df['Height'])
```

```python
upper = df["Height"].quantile(0.97)
upper
```

```python
lower = df['Height'].quantile(0.3)
lower
```

```python
new_df = df[(df['Height'] <= 73.46) & (df['Height'] >= 64.07)]
```

```python
new_df['Height'].describe()
```

```python
sns.distplot(new_df['Height'])
```

```python
sns.boxplot(x = df['Height'])
```

```python

# Capping --> Winsorization
df['Height'] = np.where(df['Height'] >= upper,
        upper,
        np.where(df['Height'] <= lower,
        lower,
        df['Height']))
```
