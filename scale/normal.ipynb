```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malica acid ']
df.head(3)
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malica acid '])
```

```python
color_dict = {1:'red', 3: 'red', 2: 'blue'}
sns.scatterplot(x = df['Alcohol'], y = df['Malica acid '] , hue = df['Class label'], palette=color_dict)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test  = train_test_split(df.drop('Class label', axis=1),df['Class label'], test_size=0.3, random_state=0)

x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Fit the scaler to the train set , it will learn the parameters 
scaler.fit(x_train) ## Fit karr diya 

x_train_Scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_Scaled = pd.DataFrame(x_train_Scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_train.columns)
```

```python
x_train_Scaled
x_test_scaled.head(3)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_Scaled.describe(),1)
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2 , figsize = (12,5))

ax1.scatter(x_train['Alcohol'], x_train['Malica acid '],c = y_train)
ax1.set_title('Before')
ax2.scatter(x_train_Scaled['Alcohol'], x_train_Scaled['Malica acid '],c = y_train)
ax2.set_title('After')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax=ax1)
sns.kdeplot(x_train['Malica acid '], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_Scaled['Alcohol'], ax=ax2)
sns.kdeplot(x_train_Scaled['Malica acid '], ax=ax1)
plt.show()
```
